```python
data = pd.read_csv('cleaned_items_with_metadata.csv')
queries_df = pd.read_csv('./data/queries.csv')

item_texts = data["full_text"].tolist()
query_texts = queries_df["search_term_pt"].tolist()
```

```python
from google import genai

client = genai.Client(api_key="XXXXXXXXXXXXXXX")

def embed_texts(texts):
    result = client.models.embed_content(
        model="text-embedding-004",
        contents=texts)
    return result.embeddings

```

```python
result = client.models.embed_content(
        model="gemini-embedding-exp-03-07",
        contents="Pizza de Frango com Catupiry é uma deliciosa combinação de frango desfiado, catupiry cremoso e uma massa crocante. É uma das pizzas mais populares no Brasil, especialmente em São Paulo. A mistura do frango com o catupiry cria um sabor único e irresistível, tornando-a uma escolha favorita entre os amantes de pizza.")

print((result.embeddings[0].values))
```

```python
import numpy as np
import time
from tqdm import tqdm
import numpy as np

# Initialize an empty list for item embeddings
item_embeddings = []

# Process item texts with tqdm progress bar
for i, text in tqdm(enumerate(item_texts), total=len(item_texts), desc="Embedding items"):
    emb = embed_texts(text)
    item_embeddings.append(np.array(emb[0].values))


# Convert to numpy and save
item_embeddings = np.array(item_embeddings)
np.save("item_embeddings_gemini_small.npy", item_embeddings)
```
